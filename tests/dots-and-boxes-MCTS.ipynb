```python
from core.implementations.MCTS import MCTS
from applications.dots_and_boxes.game_state import GameState
```

# Example 1: Two MCTS agents playing each other, mcts_1 with 1000 simulations, mcts_2 with 100 simulations.

```python
game = GameState()
print(game)
mcts_1 = MCTS(game)
mcts_2 = MCTS(game)
turn = 0
while not game.is_terminal():
    agent, num_iters = (mcts_1, 1000) if turn%2==0 else (mcts_2, 100)
    action  = agent.__call__(num_iters)
    print(action)
    game = game.apply_action(action)
    mcts_1.update_root([action])
    mcts_2.update_root([action])
    print(game)
    turn += 1
game.print_winner_and_scores()
    
```

# Example 2: Have mcts_1 and mcts_2 as above play 10 games; we expect the majority to be won by mcts_1 due to higher number of simulations

```python
game_results = {'mcts_1':0, 'draws':0, 'mcts_2':0}

for num_game in range(10):
    game = GameState()
    mcts_1 = MCTS(game)
    mcts_2 = MCTS(game)
    turn = 0
    while not game.is_terminal():
        agent, num_iters = (mcts_1, 1000) if turn%2==0 else (mcts_2, 100)
        action  = agent.__call__(num_iters)
        game = game.apply_action(action)
        mcts_1.update_root([action])
        mcts_2.update_root([action])
        turn += 1
    print(f"Result of game {num_game+1}:")
    print(game)
    game.print_winner_and_scores()
    #update game results
    if game.is_winner() == 1:
        game_results['mcts_1']+=1
    elif game.is_winner() == -1:
        game_results['mcts_2']+=1
    elif game.is_winner() == 0:
        game_results['draws']+=1
print("The overall results are:")
print(game_results)
    
```
